```python
import pandas as pd
import numpy as np





# Calculate the IQR for all features
q1 = uber_dataset.quantile(0.25)
q3 = uber_dataset.quantile(0.75)
iqr = q3 - q1

# Determine the lower and upper bounds for all features
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identify the outliers in the DataFrame
outliers = uber_dataset[(uber_dataset < lower_bound) | (uber_dataset > upper_bound)].any(axis=1)

# Identify the columns with potential outliers
outlier_cols = ['price', 'distance','surge_multiplier', 'latitude',
                'temperature','apparentTemperature','short_summary','precipIntensity','precipProbability','windGust','visibility',
                'temperatureHigh','apparentTemperatureHigh','apparentTemperatureLow','dewPoint','uvIndex','visibility.1',
                'temperatureMin','temperatureMax','apparentTemperatureMin','apparentTemperatureMax']

# Replace outliers with the upper and lower bounds
for col in outlier_cols:
    q1 = uber_dataset[col].quantile(0.25)
    q3 = uber_dataset[col].quantile(0.75)
    iqr = q3 - q1
    upper_bound = q3 + 1.5*iqr
    lower_bound = q1 - 1.5*iqr
    uber_dataset[col] = np.where(uber_dataset[col] > upper_bound, upper_bound, uber_dataset[col])
    uber_dataset[col] = np.where(uber_dataset[col] < lower_bound, lower_bound, uber_dataset[col])

```
